```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
CSV_PATH = "clean_house_dataset.csv"
df = pd.read_csv(CSV_PATH)
```

```python
# Load Dataset
df.head()
```

```python
df.shape
```

```python
x = df.drop(columns=['Price', 'LogPrice'])
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
lr_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
```

```python
lr_prediction = lr_model.predict(X_test)
rf_prediction = rf_model.predict(X_test)
```

```python
def print_metrics(name, y, y_predict):
    r2 = r2_score(y, y_predict)
    mae = mean_absolute_error(y, y_predict)
    mse = mean_squared_error(y, y_predict)
    rmse = np.sqrt(mse)
    print(f"Prediction of: {name}")
    print(f"R2: {r2:.3f}")
    print(f"MAE: {mae:,.0f}")
    print(f"MSE: {mse:,.0f}")
    print(f"RMSE: {rmse:,.0f}")
```

```python
i = 4
x_one_df = X_train.iloc[[i]]
y_true = y_test.iloc[i]
```

```python
pred_lr_one = float(lr_model.predict(x_one_df)[0])
pred_rf_one = float(rf_model.predict(x_one_df)[0])
```

```python
print_metrics("Linear Regression", y_test, lr_prediction)
print_metrics("Random Forest", y_test, rf_prediction)
```

```python
print("Single Row Sanity Check")
print(f"Actual price ${y_true:,.0f}")
print(f"LR prediction: ${pred_lr_one:,.0f}")
print(f"RF prediction: ${pred_rf_one:,.0f}")
```
