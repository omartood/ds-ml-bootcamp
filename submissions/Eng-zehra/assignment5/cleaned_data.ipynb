```python
import pandas as pd 
#Splits your dataset into training data and testing data.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#how well your model is performing.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df=pd.read_csv("Clean_Card_Price.csv")
```

```python
df.head()
```

```python
for col in ["Location_City", "Location_Rural", "Location_Suburb"]:
    df[col] = df[col].astype(int)
# To see the change, you can print the head of the DataFrame
print(df.head(20))
```

```python
x =df.drop("Price",axis=1)
y= df["Price"]
```

```python
x_train,x_test,y_train ,y_test = train_test_split(x,y,test_size=0.2 ,random_state =42)
```

```python
# initial first
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred =model.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print(f"mean absolute error (MAE):{mae}")
print(f"mean squared error (MAE):{mse}")
print(f"r2 squared rror (R2):{r2}")
```

```python
df.to_csv("cleaned_data.csv",index=False)
```

```python
row_to_check = 5
single_row_data = x_test.iloc[[row_to_check]]  

actual_price = y_test.iloc[row_to_check]

predicted_price_model = model.predict(single_row_data)
print("Actual price:", actual_price)
print("Predicted price (Model):", predicted_price_model[0])
```
