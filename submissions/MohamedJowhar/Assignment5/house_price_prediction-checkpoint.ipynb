```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
Path="clean_house_l5_dataset.csv"
df = pd.read_csv(Path)
x=df.drop(columns=['Price',"LogPrice"])
y=df["Price"]
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr_pr=lr.predict(x_test)
```

```python
rf=RandomForestRegressor(n_estimators=100,random_state=42 )
rf.fit(x_train,y_train)
```

```python
lf_pr=rf.predict(x_test)
# print(y_test.values[:5])
```

```python
def matrc_predict(name,actual_value,Predctedvalue):
    r2 =r2_score(actual_value,Predctedvalue)

    mae=mean_absolute_error(actual_value,Predctedvalue)
    mse=mean_squared_error(actual_value,Predctedvalue)
    rmse=np.sqrt(mse)
    print(f"\n{name} Performance:")
    print(f"  R²   : {r2:.3f}")        
    print(f"  MAE  : {mae:,.0f}")       
    print(f"  MSE  : {mse:,.0f}")        
    print(f"  RMSE : {rmse:,.0f}")

matrc_predict("Lr",y_test,lr_pr)
matrc_predict("fr",y_test,lf_pr)
```

```python
i=10
x_one=x_test.iloc[[i]]
y_true   = y_test.iloc[i]   
p_lr_one = float(lr.predict(x_one)[0])
p_rf_one = float(rf.predict(x_one)[0])
print(f"  Actual Price: ${y_true:,.0f}")
print(f"  lr     : ${p_lr_one:,.0f}")
print(f"  rf     : ${p_rf_one:,.0f}")
```
